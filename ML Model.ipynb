### 1. Check for the outliers in the “amount_in_usd” column and replace the outliers with appropriate values, discussed in the sessions.

```python
#checking for outliers using InterQuartile Range
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 2 * IQR
upper_bound = Q3 + 2 * IQR
print(f'Lower bound: {lower_bound}')
print(f'Upper bound: {upper_bound}')
len(df[(df['amount_in_usd'] > upper_bound) | (df['amount_in_usd'] < lower_bound)])
```

```python
#replacing the outliers with median value
median_val = np.median(df['amount_in_usd'])
df['amount_in_usd'] = df['amount_in_usd'].apply(lambda x: median_val if x < lower_bound or x > upper_bound else x)
```

```python
#after replacing, checking if any outliers left
len(df[(df['amount_in_usd'] > upper_bound) | (df['amount_in_usd'] < lower_bound)])
```

```python
def difference_in_days(melt, lags, ffday, customer_id_col, net_amount_col):
    for i in range(ffday, lags+1):
        melt['Last-'+str(i)+'day_Sales'] = melt.groupby([customer_id_col])[net_amount_col].shift(i)
    melt = melt.reset_index(drop = True)
    melt = melt.fillna(0)
    return melt
```

### 1. Modify the dataset to pass into any type of machine learning models. 

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.svm import LinearSVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('amount_in_usd', axis=1)
```

```python
y = df['amount_in_usd']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

### 2. Try different machine learning models

```python
r2_vals = {}
rmse_vals = {}
mae_vals = {}
```

```python
def run_regression_models():
    global X_train, y_train, X_test, y_test
    models = {
        "LinearRegression": LinearRegression(),
        "Ridge": Ridge(),
        "Lasso": Lasso(),
        "LinearSVR": LinearSVR(),
        "DecisionTree": DecisionTreeRegressor(),
        "RandomForest": RandomForestRegressor(),
        "AdaBoost": AdaBoostRegressor(),
        "XGBoost": XGBRegressor()
    }
    global r2_scores, rmse_scores, mae_scores
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        r2_vals[model_name] = r2_score(y_test, y_pred)
        rmse_vals[model_name] = mean_squared_error(y_test, y_pred, squared=False)
        mae_vals[model_name] = mean_absolute_error(y_test, y_pred)
        print(f'{model_name}, r2 = {r2_vals[model_name]}, rmse = {rmse_vals[model_name]}, mae = {mae_vals[model_name]}')
```

```python
run_regression_models()
```

### 3. Perform MSE, RMSE, R-Square

```python
sorted_r2 = sorted(r2_vals.items(), key=lambda x: x[1], reverse=True)
print("R2 Scores:")
for model_name, score in sorted_r2:
    print(f"{model_name}: {score}")
```

```python
sorted_rmse = sorted(rmse_vals.items(), key=lambda x: x[1])
print("\nRMSE Scores:")
for model_name, score in sorted_rmse:
    print(f"{model_name}: {score}")
```

```python
sorted_mae = sorted(mae_vals.items(), key=lambda x: x[1])
print("\nMAE Scores:")
for model_name, score in sorted_mae:
    print(f"{model_name}: {score}")
```

### 6. Perform Hyperparameter tuning, select best hyperparameters by using appropriate algorithms

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgparam_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.01, 0.001, None],
    'max_depth': [3, 5, 7, None],
    'subsample': [0.8, 0.9, 1.0]
}
```

```python
xgb = XGBRegressor()
random_search = RandomizedSearchCV(estimator=xgb, param_distributions=xgparam_grid, n_iter=10, scoring='r2', cv=4)
random_search.fit(X, y)
results = random_search.cv_results_
for mean_score, params in zip(results['mean_test_score'], results['params']):
    print(f"Score: {mean_score:.4f} with Params: {params}")
print("\nBest Parameters: ", random_search.best_params_)
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("\nR2 Score: ", r2)
```

```python
#training the model with the Best Parameters
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

model = xgb.XGBRegressor(subsample=0.9, n_estimators=300, max_depth=None, learning_rate=0.1)

model.fit(X_train, y_train)

score = model.score(X_test, y_test)
print("Model score: ", score)
```
